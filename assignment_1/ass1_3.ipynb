```python
from sklearn.model_selection import train_test_split
import random 
# Loading and exploring the dataset
from sklearn.datasets import load_iris
irisData = load_iris()
print(irisData.data) 
print(irisData.target) 
print(irisData.feature_names)
print(irisData.target_names)
```

```python
# Standardizing the data attributes
# We standardize because the data is varying in scale and logistic regression does make assumptions about
# the data having a Gaussian distribution.
from sklearn import preprocessing
irisData.data = preprocessing.scale(irisData.data) # Scales to an average of 0, with a standard deviation of 1
print(irisData.data)
```

```python
# Splitting the dataset into testing and training set
X_train,X_test,Y_train,Y_test = train_test_split( irisData.data,irisData.target,
                                                 test_size=0.3,
                                                 random_state =random.seed())
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Logistic regression is by default meant for binary classification, two classes, however we can use multi class logistic regression here
from sklearn.linear_model import LogisticRegression

lr= LogisticRegression(multi_class='ovr')
lr=lr.fit(X_train, Y_train)   # Here we train the model
```

```python
y_pred_train = lr.predict(X_train) # Getting the predicted values based on X_train data
y_pred_test = lr.predict(X_test)   # Getting predicted y - values based on the unseen test data
```

```python
from sklearn.metrics import accuracy_score   # Measuring the accuracy of the model. It seems as if the model performed better on
print(accuracy_score(Y_train,y_pred_train))  # the unseen test data. This could very easily be because the sample data is very small, or by
print(accuracy_score(Y_test,y_pred_test))    # chance that some of the more "easy" patterns were in the test data
```

```python
# Support vector machine with polynomial kernel, degree 3, for classification of three distinct classes in the iris datatset
from sklearn import svm
mch = svm.SVC(kernel='poly', degree=3, C=1,decision_function_shape='ovo').fit(X_train, Y_train)
```

```python
y_pred_train = mch.predict(X_train) # Predicting the y value based on the training data
y_pred_test = mch.predict(X_test)   # Predicting the y-value based on the test data
```

```python
print(accuracy_score(Y_train,y_pred_train))  # From the accuracy scores we can see that SVM actually fared a little better than logistic 
print(accuracy_score(Y_test,y_pred_test))    # regression, but still quite similar results. We also again see the phenomenon of the model 
                                             # scoring better on the testing data. Again this may be because of the small sample size, or
                                             # maybe easier patterns to recognize in the test data
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Y = to_categorical(irisData['target']) # Converting the target data to binary vectors [1,0,0] is 0 in classification, [0,1,0] is 1 and so
#print(Y)
# Doing new test train split because of this conversion of data
X_train, X_test, Y_train, Y_test = train_test_split(irisData.data, Y, test_size=0.3, random_state =random.seed())
```

```python
model = models.Sequential() 
model.add(layers.Dense(10, input_dim=X_train.shape[1], activation='relu')) # Not using dropout here, which you would normally use to prevent overfitting
model.add(layers.Dense(3, activation='softmax')) # Three outputs, activation is softmax because it is a multi class classification problem
```

```python
# Compilation of our model, using categorical crossentropy as loss because it is a multiclass problem
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# Fitting the model
results = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=200, batch_size=10)
# Batch size controls number of training samples to work thorugh before the models internal parameters are updated
```

```python
print("Test-Accuracy:", np.mean(results.history["val_accuracy"]))
losses = results.history['loss']
val_losses = results.history['val_loss']
# The accuracy of our neural network was much lower than the other models. This is likely because neural network architectures
# perform the best on large amounts of data, whereas traditional machine learning models can perform good on small amounts of data.
# In this case with the Iris dataset, there aren't really lots of data, and this is why the other traditional machine learning models
# outperform the neural networks here
```

```python
plt.figure()
history_dict = results.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, (len(history_dict['loss']) + 1))
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# As we can see from this graph, we notice that the validation loss is bigger than the training loss. This could mean that our model is 
# overfitting. This makes perfect sense, as I didn't add any dropout to the neural network

# Lastly I want to mention normalization and why we do this to our data.
# If a feature in the dataset is big in scale compared to the others, then this big scaled feature becomes dominating, and as a result of 
# that, predictions from the neural network will not be accurate. Normalizing this data between 0 and 1 will mean that the 
# scaled feature will not dominate and ruin the accuracy of the model
```
